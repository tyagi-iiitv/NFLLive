```python
import numpy as np
import matplotlib.pyplot as plt
import pylab
import warnings
warnings.filterwarnings("ignore")
```

```python
st = 100
```

```python
num = 101
Pf = np.linspace(0, 100, num) / 100
Pu = 1 - Pf
```

```python
def MLf (m):
   return -(st * Pf * (1 + m)) / (1 - Pf * (1 + m));
def MLu (m):
   return (st * (1 - (1-Pf)*(1 + m))) / ((1-Pf) * (1 + m));
```

```python
m0 = 0
m1 = 0.1
m2 = 0.2
```

```python
offset = 30
```

```python
plt.figure(figsize = (16, 10))
m0_u = plt.plot(Pf[offset : num-offset], MLu(m0)[offset : num-offset], 'g--', label = 'm = '+str(m0)+', Underdog')
m1_u = plt.plot(Pf[offset : num-offset], MLu(m1)[offset : num-offset], 'r--', label = 'm = '+str(m1)+', Underdog')
m2_u = plt.plot(Pf[offset : num-offset], MLu(m2)[offset : num-offset], 'b--', label = 'm = '+str(m2)+', Underdog')
m0_f = plt.plot(Pf[offset : num-offset], MLf(m0)[offset : num-offset], 'g', label = 'm = '+str(m0)+', Favorite')
m1_f = plt.plot(Pf[offset : num-offset], MLf(m1)[offset : num-offset], 'r', label = 'm = '+str(m1)+', Favorite')
m2_f = plt.plot(Pf[offset : num-offset], MLf(m2)[offset : num-offset], 'b', label = 'm = '+str(m2)+', Favorite')
plt.text(Pf[num-offset-1], MLf(m0)[num-offset-1], 'm = '+str(m0), horizontalalignment='center', fontsize = 12)
plt.text(Pf[num-offset-1], MLu(m0)[num-offset-1], 'm = '+str(m0), horizontalalignment='center', fontsize = 12)
plt.text(Pf[num-offset-1], MLf(m1)[num-offset-1], 'm = '+str(m1), horizontalalignment='center', fontsize = 12)
plt.text(Pf[num-offset-1], MLu(m1)[num-offset-1], 'm = '+str(m1), horizontalalignment='center', fontsize = 12)
plt.text(Pf[num-offset-1], MLf(m2)[num-offset-1], 'm = '+str(m2), horizontalalignment='center', fontsize = 12)
plt.text(Pf[num-offset-1], MLu(m2)[num-offset-1], 'm = '+str(m2), horizontalalignment='center', fontsize = 12)
plt.text(Pf[offset+1], MLu(m0)[round(offset*1.4)], 'Underdogs', horizontalalignment='center', fontsize = 12)
plt.text(Pf[offset+1], MLf(m2)[round(offset*1.4)], 'Favourites', horizontalalignment='center', fontsize = 12)
plt.legend(fontsize = 14)
plt.xlabel('Win probability', fontsize = 16)
plt.ylabel('Moneyline', fontsize = 16)
pylab.xlim([(offset-1)*0.01, (num-offset)*0.01+0.01])
plt.show()
```

```python
plt.figure(figsize = (16, 10))
m2_f = plt.plot(Pf[offset : num-offset], -MLf(m2)[offset : num-offset], 'b', label = 'm = '+str(m2)+', -Favorite')
m1_f = plt.plot(Pf[offset : num-offset], -MLf(m1)[offset : num-offset], 'r', label = 'm = '+str(m1)+', -Favorite')
m0_pl = plt.plot(Pf[offset : num-offset], -MLf(m0)[offset : num-offset], 'g', label = 'm = '+str(m0)+', Favorite = -Underdog')
m1_u = plt.plot(Pf[offset : num-offset], MLu(m1)[offset : num-offset], 'r--', label = 'm = '+str(m1)+', Underdog')
m2_u = plt.plot(Pf[offset : num-offset], MLu(m2)[offset : num-offset], 'b--', label = 'm = '+str(m2)+', Underdog')
plt.text(Pf[num-offset-1], -MLf(m0)[num-offset-1], 'Favorite = -Underdog', horizontalalignment='center', fontsize = 12)
plt.text(Pf[num-offset-1], -MLf(m1)[num-offset-1], 'Favorite (-)', horizontalalignment='center', fontsize = 12)
plt.text(Pf[num-offset-1], MLu(m1)[num-offset-1], 'Underdog', horizontalalignment='center', fontsize = 12)
plt.text(Pf[num-offset-1], -MLf(m2)[num-offset-1], 'Favorite (-)', horizontalalignment='center', fontsize = 12)
plt.text(Pf[num-offset-1], MLu(m2)[num-offset-1], 'Underdog', horizontalalignment='center', fontsize = 12)
plt.legend(fontsize = 14)
plt.xlabel('Win probability', fontsize = 16)
plt.ylabel('Moneyline', fontsize = 16)
pylab.xlim([(offset-1)*0.01, (num-offset)*0.01+0.01])
plt.show()
```
